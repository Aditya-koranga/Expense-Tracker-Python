# Personal Expense tracker
## This project tracks Personal expenses and calculate

- Total Expense
- Remaining Budget
- Budget Left after each Expense
(*Built Using python in Jupyter Notebook*)

```python
# Expense Tracker 
Total_bud = 30000

exp1 = 3500 #(Coursefee)
exp2 = 2300 #(1 Emi)
exp3 = 2000 #(Bike Service)
exp4 = 14000 #(Cash out)
exp5 = 2000 #(tution exp)
exp6 = 1452 + 1735 + 110 #(Miscleneus)

Total_exp = exp1 + exp2 + exp3 + exp4 + exp5 + exp6

Budget_Left = Total_bud - Total_exp

available_money1 = Total_bud - exp1 
available_money2 = available_money1 - exp2
available_money3 = available_money2 - exp3
available_money4 = available_money3 - exp4
available_money5 = available_money4 - exp5
available_money6 = available_money5 - exp6

print(f'Budget left after first exp {available_money1}')
print(f'Budget left afyter second exp {available_money2}')
print(f'Budget left after third exp {available_money3}')
print(f'Budget left after fourth exp {available_money4}')
print(f'Budget left after fifth exp {available_money5}')
print(f'Budget left after sixth exp {available_money6}')
print(f'Total expenses are {Total_exp}')
print(f'Budget left in hand {Budget_Left}')
```
